Pandas is a package for data analysis in Python.  The main data struture in Pandas is the DataFrame.


```python
# import
import pandas as pd
```

```python
# create a dictionary with data
data = {'School': ['USC', 'USC', 'USC', 'USC', 'USC', 'UGA', 'UGA', 'UGA', 'UGA'] ,
        'Year': [2014, 2015, 2016, 2017, 2018, 2015, 2016, 2017, 2018],
        'Wins': [7, 3, 6, 9, 7, 10, 8, 13, 11]}
data
```

```python
# Load these data into a Pandas DataFrame
df = pd.DataFrame(data)
df
```

```python
df['Wins'].describe()
```

# Reading data into Pandas

Now let's read in larger dataset from a csv file posted to the web.  With richer data, we need to explore to see ranges of values, columns of varialbes, etc.

```python
kisa_data = pd.read_csv(
    'https://www.kauffman.org/-/media/kauffman_org/kauffman-index/archive-data/kieadata15.csv?la=en')
```

```python
kisa_data.head(n=5)
```

```python
kisa_data.keys()
```

```python
kisa_data.describe()
```

```python
kisa_data[['faminc', 'age']].groupby('age').mean()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
kisa_data[['age', 'faminc']].groupby('age').mean().plot.bar(y='faminc')
```
